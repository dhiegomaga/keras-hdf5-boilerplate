```python
# Imports, variables

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import csv
import os
#import cv2
import glob
import random
import pickle
from tqdm import tqdm

DATADIR = "Dataset/"
CSV_FILE = DATADIR+"labels.csv"
CSV_FIELD_COUNT = 2
STATE_FILE = 'state.pickle'
CATEGORIES = ["Dog", "Cat"]
IMG_SIZE = 100
LABELS_PATH = "Y.pickle"
FEATURES_PATH = "X.pickle"
```

```python
brandNames = {}
records = []
```

```python
# Load brand names

fi = open(STATE_FILE, 'rb')
state = pickle.load(fi)
fi.close()
brandNames=state['brandNames']
print(brandNames)
```

```python
# Load CSV

with open(CSV_FILE) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    next(csv_reader) 
    for record in csv_reader:
        if(len(record) != CSV_FIELD_COUNT):
            print("Wrong CSV format")
            break
        records.append(record)
```

```python
# Visualize data sample

index = 1109

file, label = records[index]
image = mpimg.imread(file)
plt.imshow(image)
print(brandNames[int(label)])
```
